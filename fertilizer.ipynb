```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import pickle

# डेटा लोड करा
df = pd.read_csv('data_core.csv')

# शब्दांना नंबरमध्ये बदलण्यासाठी एन्कोडर्स
le_soil = LabelEncoder()
df['Soil Type'] = le_soil.fit_transform(df['Soil Type'])

le_crop = LabelEncoder()
df['Crop Type'] = le_crop.fit_transform(df['Crop Type'])

# फीचर्स आणि टारगेट (CSV मधील स्पेलिंग 'Temparature' असे आहे)
X = df.drop(['Fertilizer Name', 'Moisture'], axis=1) 
y = df['Fertilizer Name']

# मॉडेल ट्रेनिंग
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# सर्व फाईल्स सेव्ह करा
pickle.dump(model, open('fert_model.pkl', 'wb'))
pickle.dump(le_soil, open('le_soil.pkl', 'wb'))
pickle.dump(le_crop, open('le_crop.pkl', 'wb'))

print("Fertilizer Model Saved!")

```
